# 2. Constructing the data set

Iterate through the csv files from the data/points folder and collect the contents in a common dataframe


```python
import pandas as pd
import glob

file_pattern = '*.csv'

# Get a list of CSV file paths using glob
csv_files = glob.glob('data/points/' + file_pattern)
dataframes = []

# Iterate over the CSV files and read them one by one
for file in csv_files:
    df = pd.read_csv(file)
    dataframes.append(df)

df_points = pd.concat(dataframes, ignore_index=True)
df_points = df_points[["match_id","SetNo","GameNo","PointNumber","PointWinner","PointServer","Speed_KMH","P1Score","P2Score","P1Ace","P2Ace","P1Winner","P2Winner","P1DoubleFault","P2DoubleFault","P1UnfErr","P2UnfErr","ServeNumber","RallyCount"]]
df_points = df_points[((df_points['PointNumber'] != '0X') & (df_points['PointNumber'] != '0Y'))]
df_points.shape
df_points.dtypes
df_points

```

Iterate through the csv files from the data/matches folder to construct dataframes for men and women separately.

*   Extract the tournament's surface base on the name of the file
*   Check the match num to append the match to the correct datfarame (men/women)
*   Shorten the name of the players, to keep them in a consistent format



```python
file_pattern = '*.csv'

def shorten_name(name):
    # Split the name into a list of names
    names = name.split()
    # Get the last name
    last_name_parts = names[-1].split('-')
    last_name = last_name_parts[-1]
    # Get the first initials of all the first names
    first_initials = []
    for name in names[:-1]:
        if '-' in name:
            for part in name.split('-'):
                first_initials.append(part[0].upper())
        else:
            first_initials.append(name[0].upper())

    for part in last_name_parts[:-1]:
        first_initials.append(part[0].upper())

    first_initials = [f'{initial}.' for initial in first_initials]
    # Return the first initials followed by the last name
    return f'{" ".join(first_initials)} {last_name}'

# Get a list of CSV file paths using glob
csv_files = glob.glob('data/matches/' + file_pattern)
dataframes_men = []
dataframes_women = []

# Iterate over the CSV files and read them one by one
for file in csv_files:
    df = pd.read_csv(file)
    print(file)

   # Extract tournament name from file name
    tournament = file.split('/')[2].split('-')[1]

    # Set surface variable based on tournament name
    if tournament == 'frenchopen':
        surface = 'clay'
    elif tournament == 'wimbledon':
        surface = 'grass'
    elif tournament == 'ausopen' or tournament == 'usopen':
        surface = 'hard'
    else:
        surface = 'unknown'

    print(f'Tournament: {tournament}')
    print(f'Surface: {surface}')

    if df['match_num'].dtype == 'int64':
        men_matches = df[df['match_num'] < 2000]
        women_matches = df[df['match_num'] >= 2000]
    else:
        men_matches = df[df['match_num'].str.startswith('MS')]
        women_matches = df[df['match_num'].str.startswith('WS')]

    men_matches = men_matches.assign(surface=surface)
    women_matches = women_matches.assign(surface=surface)

    dataframes_men.append(men_matches)
    dataframes_women.append(women_matches)


df_matches_men = pd.concat(dataframes_men, ignore_index=True)
df_matches_women = pd.concat(dataframes_women, ignore_index=True)
df_matches_men = df_matches_men[["match_id","year","slam","match_num","player1","player2","nation1","nation2", "surface"]]
df_matches_women = df_matches_women[["match_id","year","slam","match_num","player1","player2","nation1","nation2", "surface"]]




# Shorten the player names

df_matches_men['player1'] = df_matches_men['player1'].apply(shorten_name)
df_matches_men['player2'] = df_matches_men['player2'].apply(shorten_name)
df_matches_women['player1'] = df_matches_women['player1'].apply(shorten_name)
df_matches_women['player2'] = df_matches_women['player2'].apply(shorten_name)

df_matches_women
```

Separate the points dataframe into men and women and merge them with the match dataframes


```python
# Men matches and points

df_men_points = pd.merge(df_matches_men,df_points)
df_men_points = df_men_points[["match_id", "year","slam","surface", "match_num","player1","player2","nation1","nation2", "SetNo","GameNo","PointNumber","PointWinner","PointServer","Speed_KMH","P1Score","P2Score","ServeNumber","RallyCount"]]
df_men_points

# Women matches and points

df_women_points = pd.merge(df_matches_women,df_points)
df_women_points = df_women_points[["match_id", "year","slam","surface", "match_num","player1","player2","nation1","nation2", "SetNo","GameNo","PointNumber","PointWinner","PointServer","Speed_KMH","P1Score","P2Score","ServeNumber","RallyCount"]]
df_women_points


```

## 3.2 IsBreakPointFeature

 Put break point and non-break points into different tables


```python
df_men_break = df_men_points[df_men_points['GameNo'] == 0]

# Fill missing or NaN values in the columns used to create the boolean mask
df_men_points['PointServer'] = df_men_points['PointServer'].fillna(0)
df_men_points['P1Score'] = df_men_points['P1Score'].fillna('')
df_men_points['P2Score'] = df_men_points['P2Score'].fillna('')

# Create boolean masks to identify break opportunities
p1_break_opportunity = (
    (df_men_points['PointServer'] == 2) &
    (
        ((df_men_points['P1Score'] == '40') & (df_men_points['P2Score'] == '0')) |
        ((df_men_points['P1Score'] == '40') & (df_men_points['P2Score'] == '15')) |
        ((df_men_points['P1Score'] == '40') & (df_men_points['P2Score'] == '30')) |
        ((df_men_points['P1Score'] == 'AD') & (df_men_points['P2Score'] == '40'))
    )
)

p2_break_opportunity = (
    (df_men_points['PointServer'] == 1) &
    (
        ((df_men_points['P1Score'] == '0') & (df_men_points['P2Score'] == '40')) |
        ((df_men_points['P1Score'] == '15') & (df_men_points['P2Score'] == '40')) |
        ((df_men_points['P1Score'] == '30') & (df_men_points['P2Score'] == '40')) |
        ((df_men_points['P1Score'] == '40') & (df_men_points['P2Score'] == 'AD'))
    )
)

break_opportunity = p1_break_opportunity | p2_break_opportunity

# Create a boolean mask to identify the next point after each break opportunity
next_point = break_opportunity.shift(+1).fillna(False)

# Add the rows corresponding to breakpoints to the df_men_break dataframe
breakpoints = df_men_points[next_point]
df_men_break = pd.concat([df_men_break, breakpoints])



df_men_break
```

## 3.3 PLAYER_DIFF

Get points won be each player on serve and return

```python
import numpy as np

def calc_points(df, player_col, server_val, winner_val):
    grouped = df.groupby([player_col])
    serve_points_played = grouped.apply(lambda x: (x['PointServer'] == server_val).sum())
    serve_points_won = grouped.apply(lambda x: ((x['PointServer'] == server_val) & (x['PointWinner'] == winner_val)).sum())
    return_points_played = grouped.apply(lambda x: (x['PointServer'] == (3 - server_val)).sum())
    return_points_won = grouped.apply(lambda x: ((x['PointServer'] != server_val) & (x['PointWinner'] == winner_val)).sum())
    new_df = pd.DataFrame({
        'Player': serve_points_played.index.get_level_values(player_col),
        'ServePointsPlayed': serve_points_played.values,
        'ServePointsWon': serve_points_won.values,
        'ReturnPointsPlayed': return_points_played.values,
        'ReturnPointsWon': return_points_won.values
    })
    return new_df

def calc_df(df, surface=None):
    # Filter input DataFrame by surface type if surface is not None
    if surface is not None:
        df = df[df['surface'] == surface]

    new_df_p1 = calc_points(df, 'player1', 1, 1)
    new_df_p2 = calc_points(df, 'player2', 2, 2)
    df_player_points_won = pd.concat([new_df_p1, new_df_p2], ignore_index=True, sort=False)

    grouped = df_player_points_won.groupby(['Player'])
    serve_points_played = grouped['ServePointsPlayed'].sum()
    serve_points_won = grouped['ServePointsWon'].sum()
    return_points_played = grouped['ReturnPointsPlayed'].sum()
    return_points_won = grouped['ReturnPointsWon'].sum()

    service_win_percentage = np.divide(serve_points_won.values, serve_points_played.values, out=np.zeros_like(serve_points_won.values, dtype=float), where=serve_points_played.values!=0)
    return_win_percentage = np.divide(return_points_won.values, return_points_played.values, out=np.zeros_like(return_points_won.values, dtype=float), where=return_points_played.values!=0)

    df_player_points_won = pd.DataFrame({
        'Player': serve_points_played.index.get_level_values('Player'),
        'ServePointsPlayed': serve_points_played.values,
        'ServePointsWon': serve_points_won.values,
        'ServiceWinPercentage': service_win_percentage,
        'ReturnPointsPlayed': return_points_played.values,
        'ReturnPointsWon': return_points_won.values,
        'ReturnWinPercentage': return_win_percentage
    })

    total_serve_points_played = df_player_points_won['ServePointsPlayed'].sum()
    total_serve_points_won = df_player_points_won['ServePointsWon'].sum()
    avg_serve_win_percentage = total_serve_points_won / total_serve_points_played

    total_return_points_played = df_player_points_won['ReturnPointsPlayed'].sum()
    total_return_points_won = df_player_points_won['ReturnPointsWon'].sum()
    avg_return_win_percentage = total_return_points_won / total_return_points_played

    return df_player_points_won, avg_serve_win_percentage, avg_return_win_percentage

men_avg_serve_win_pct_surf_dict = {}
men_avg_return_win_pct_surf_dict = {}
women_avg_serve_win_pct_surf_dict = {}
women_avg_return_win_pct_surf_dict = {}
df_men_res_surf_dict = {}
df_women_res_surf_dict = {}

df_men_res, men_avg_serve_win_pct, men_avg_return_win_pct = calc_df(df_men_points)
df_men_res_break, men_avg_serve_win_pct_break, men_avg_return_win_pct_break = calc_df(df_men_break)
df_women_res, women_avg_serve_win_pct, women_avg_return_win_pct = calc_df(df_women_points)

# Compute dataframes for each surface separately
surfaces = ['clay', 'grass', 'hard']

for surface in surfaces:
    print(f'Calculating results for {surface} surface...')


    df_men_res_surf, men_avg_serve_win_pct_surf, men_avg_return_win_pct_surf = calc_df(df_men_points, surface)
    df_women_res_surf, women_avg_serve_win_pct_surf, women_avg_return_win_pct_surf = calc_df(df_women_points, surface)

    men_avg_serve_win_pct_surf_dict[surface] = men_avg_serve_win_pct_surf
    men_avg_return_win_pct_surf_dict[surface] = men_avg_return_win_pct_surf
    women_avg_serve_win_pct_surf_dict[surface] = women_avg_serve_win_pct_surf
    women_avg_return_win_pct_surf_dict[surface] = women_avg_return_win_pct_surf
    df_men_res_surf_dict[surface] = df_men_res_surf
    df_women_res_surf_dict[surface] = df_women_res_surf

print(men_avg_serve_win_pct)
print(men_avg_return_win_pct)
df_men_res
```

Caclulate the PLAYER_DIFF feature using the win percentages

```python
def calc_player_diff(df, avg_serve_win_pct, avg_return_win_pct):
    df['key'] = 1
    df_merged = pd.merge(df, df, on='key', suffixes=('', '_y'))
    df_merged = df_merged[df_merged['Player'] != df_merged['Player_y']]
    df_merged['PLAYER_DIFF'] = (df_merged['ServiceWinPercentage'] - avg_serve_win_pct) - (df_merged['ReturnWinPercentage_y'] - avg_return_win_pct)
    df_player_diff = df_merged[['Player', 'Player_y', 'ServiceWinPercentage', 'ReturnWinPercentage_y', 'PLAYER_DIFF']]
    df_player_diff.columns = ['Player1', 'Player2', 'Player1ServeWin', 'Player2ReturnWin', 'PLAYER_DIFF']
    return df_player_diff

df_men_player_diff = calc_player_diff(df_men_res, men_avg_serve_win_pct, men_avg_return_win_pct)
df_women_player_diff = calc_player_diff(df_women_res, women_avg_serve_win_pct, women_avg_return_win_pct)

df_men_player_diff_surf_dict = {}
df_women_player_diff_surf_dict = {}

for surface in surfaces:
    df_men_res_surf = df_men_res_surf_dict[surface]
    men_avg_serve_win_pct_surf = men_avg_serve_win_pct_surf_dict[surface]
    men_avg_return_win_pct_surf = men_avg_return_win_pct_surf_dict[surface]

    df_women_res_surf = df_women_res_surf_dict[surface]
    women_avg_serve_win_pct_surf = women_avg_serve_win_pct_surf_dict[surface]
    women_avg_return_win_pct_surf = women_avg_return_win_pct_surf_dict[surface]

    df_men_player_diff_surf = calc_player_diff(df_men_res_surf, men_avg_serve_win_pct_surf, men_avg_return_win_pct_surf)
    df_women_player_diff_surf = calc_player_diff(df_women_res_surf, women_avg_serve_win_pct_surf, women_avg_return_win_pct_surf)

    df_men_player_diff_surf_dict[surface] = df_men_player_diff_surf
    df_women_player_diff_surf_dict[surface] = df_women_player_diff_surf

print(men_avg_serve_win_pct)
print(men_avg_return_win_pct)
df_men_player_diff
```

## 3.4 Preparing the data for model fitting


```python
def process_data(df, df_player_diff, is_break_point):
    df = df[["player1", "player2","PointWinner","PointServer", "PreviousPointWinner"]]
    df = df[(df['PointServer'] != 0) & (df['PointWinner'] != 0)]

    df_serve = pd.DataFrame()
    df_serve['Player1'] = df.apply(lambda x: x['player1'] if x['PointServer'] == 1 else x['player2'], axis=1)
    df_serve['Player2'] = df.apply(lambda x: x['player2'] if x['PointServer'] == 1 else x['player1'], axis=1)
    df_serve['PreviousPointWinner'] = df.apply(lambda x: 1 if (x['PointServer'] == x['PreviousPointWinner']) else 2 if ((x['PointServer'] + x['PreviousPointWinner']) == 3) else 0, axis=1)
    df_serve['Winner'] = df.apply(lambda x: 1 if x['PointWinner'] == x['PointServer'] else 2, axis=1)
    df_serve['IsBreakPoint'] = is_break_point

    merged_df = pd.merge(df_serve, df_player_diff[['Player1', 'Player2', 'PLAYER_DIFF']], on=['Player1', 'Player2'], how='left')

    return merged_df

men_non_break_data = process_data(df_men_non_break, df_men_player_diff, 0)
men_break_data = process_data(df_men_break, df_men_player_diff, 1)
women_non_break_data = process_data(df_women_non_break, df_women_player_diff, 0)
women_break_data = process_data(df_women_break, df_women_player_diff, 1)

men_data = pd.concat([men_non_break_data, men_break_data], ignore_index=True)
women_data = pd.concat([women_non_break_data, women_break_data], ignore_index=True)

men_data_surf = {}
women_data_surf = {}

for surface in surfaces:
    men_non_break_data_surf = process_data(df_men_non_break[df_men_non_break['surface'] == surface], df_men_player_diff_surf_dict[surface], 0)
    men_break_data_surf = process_data(df_men_break[df_men_break['surface'] == surface], df_men_player_diff_surf_dict[surface], 1)
    men_data_surf[surface] = pd.concat([men_non_break_data_surf, men_break_data_surf], ignore_index=True)

    women_non_break_data_surf = process_data(df_women_non_break[df_women_non_break['surface'] == surface], df_women_player_diff_surf_dict[surface], 0)
    women_break_data_surf = process_data(df_women_break[df_women_break['surface'] == surface], df_women_player_diff_surf_dict[surface], 1)
    women_data_surf[surface] = pd.concat([women_non_break_data_surf, women_break_data_surf], ignore_index=True)

men_data
```

## 5.3 Player cluster analysis


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load the dataframe with 'Player', 'lambda_S', and 'lambda_R' columns
data = men_lambdas

# Extract the features for clustering
features = data[['lambda_S', 'lambda_R']]

# Define the number of clusters
num_clusters = 5

# Perform K-means clustering with the chosen number of clusters
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(features)

# Get the cluster labels for each player
cluster_labels = kmeans.labels_

# Add the cluster labels to the dataframe
data['Cluster'] = cluster_labels

# Plot the players with different colors for each cluster
colors = ['red', 'blue', 'green', 'orange', 'purple']
for cluster in range(num_clusters):
    cluster_data = data[data['Cluster'] == cluster]
    plt.scatter(cluster_data['lambda_S'], cluster_data['lambda_R'], c=colors[cluster], label=f'Cluster {cluster}')


# Add labels and title to the plot
plt.xlabel('$\lambda^S$')
plt.ylabel('$\lambda^R$')
plt.title('Player Clusters')
plt.legend()

# Show the plot
plt.show()
data = data[['Player', 'lambda_S', 'lambda_R', 'Cluster']]
data
```

```python
from scipy.stats import zscore

surfaces = ['clay', 'hard', 'grass']
z_threshold = 3

for surface in surfaces:
    # Get the data for this surface
    men_lambdas_surface = men_lambdas_surf[surface]

    # Compute the z-scores for the lambda_S and lambda_R columns
    men_lambdas_surface['z_lambda_S'] = zscore(men_lambdas_surface['lambda_S'])
    men_lambdas_surface['z_lambda_R'] = zscore(men_lambdas_surface['lambda_R'])

    # Filter the data to exclude rows where the absolute z-score is greater than the threshold
    men_lambdas_filtered = men_lambdas_surface[(men_lambdas_surface['z_lambda_S'].abs() <= z_threshold) & (men_lambdas_surface['z_lambda_R'].abs() <= z_threshold)]

    # Reset the index of the men_lambdas_filtered DataFrame
    men_lambdas_filtered = men_lambdas_filtered.reset_index(drop=True)

    x = men_lambdas_filtered['lambda_S']
    y = men_lambdas_filtered['lambda_R']

    fig, ax = plt.subplots(figsize=(8, 8))
    ax.scatter(x, y)

    # add labels to points with extreme lambda values
    for i, player in enumerate(men_lambdas_filtered['Player']):
        if x[i] > x.quantile(0.90) or y[i] > y.quantile(0.80) or x[i] < x.quantile(0.05) or y[i] < y.quantile(0.05) or player == 'R. Nadal' or player == 'N. Djokovic':
            ax.annotate(player, (x[i], y[i]))

    ax.set_xlabel('lambdaS')
    ax.set_ylabel('lambdaR')
    ax.set_title(f'Scatter plot of lambdaS vs lambdaR for {surface} surface')
    plt.show()
```
